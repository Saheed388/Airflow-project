```python
import requests
import gzip
import os
import pandas as pd
from io import BytesIO
import time

# URL of the GZIP file you want to extract
gzip_url = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2021-07.csv.gz "

# Number of retries
max_retries = 3

for attempt in range(1, max_retries + 1):
    try:
        # Send an HTTP GET request to the URL
        response = requests.get(gzip_url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Create a BytesIO object from the response content
            gzip_data = BytesIO(response.content)

            # Specify the extraction directory
            extraction_path = "/c/Users/HP/Documents/Airflow-gcp/altschool-projecr"  # Change to your desired extraction directory

            try:
                # Create the extraction directory if it doesn't exist
                os.makedirs(extraction_path, exist_ok=True)

                # Open the GZIP file
                with gzip.open(gzip_data, "rb") as gzip_file:
                    # Read the GZIP contents and decode it as CSV
                    csv_content = gzip_file.read().decode("utf-8")

                # Save the CSV content as a CSV file
                csv_file_path = os.path.join(extraction_path, "green_file.csv")
                with open(csv_file_path, "w") as csv_file:
                    csv_file.write(csv_content)

                print("Extraction and CSV conversion completed.")
                break  # Successful, exit retry loop
            except Exception as e:
                print(f"Error during extraction: {str(e)}")
        else:
            print(f"Failed to retrieve the GZIP file. Status code: {response.status_code}")

    except requests.exceptions.ConnectionError:
        if attempt < max_retries:
            print(f"Retry attempt {attempt} - Connection error. Retrying in 10 seconds...")
            time.sleep(30)
        else:
            print(f"Max retries reached. Could not establish a connection.")
```
